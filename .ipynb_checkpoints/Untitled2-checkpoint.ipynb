```r
library(tidyverse)
```

```r
wiki = read_csv("WikiData.csv")
```

```r
wiki %>% 
```

```r
wiki %>% 
    group_by(settlement_type) %>%
    filter(n() > 200,lon < 0,lon>-125,lat>20) %>%
    filter(settlement_type != "US County") %>%
    ggplot() + geom_point(aes(y=lat,x=lon,color=settlement_type),size=.33) + coord_equal()
```

```r
wiki %>% 
    filter(lon < 0,lon>-125,lat>20) %>%
    filter(state=="MA") %>%
    filter(settlement_type != "US County") %>% ggplot() + geom_point(aes(y=lat,x=lon,size=`y1890`))
```
